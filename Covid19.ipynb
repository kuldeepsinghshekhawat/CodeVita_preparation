```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
import numpy as np


df=pd.read_csv("covid_19_india.csv",parse_dates=["Date"],dayfirst=True)
df.head()
```

```python
df=df[["Date","State/UnionTerritory","Cured","Deaths","Confirmed"]]
```

```python
df.columns=["date","state","cured","deaths","confirmed"]
```

```python
df.head()
```

```python
df.tail()
```

```python
today=df[df.date=='2020-07-30']
```

```python
today
```

```python
max_confirmed_cases=today.sort_values(by="confirmed",ascending=True)
```

```python
max_confirmed_cases
```

```python
top_state_confirmed=max_confirmed_cases[0:5]
top_state_confirmed
```

```python
sns.set(rc={'figure.figsize':(15,10)})
sns.barplot(x='state',y='confirmed',data=top_state_confirmed,hue='state')
plt.show()
```

```python
max_death_cases=today.sort_values(by='deaths',ascending=False)
max_death_cases
top_death_cases=max_death_cases[0:10]
top_death_cases
```

```python
sns.set(rc={'figure.figsize':(15,10)})
sns.barplot(x='state',y='deaths',data=top_death_cases,hue='state')
plt.show()
```

```python
max_cured_cases=today.sort_values(by="cured",ascending=False)
max_cured_cases
top_cured_cases=max_cured_cases[0:10]
top_cured_cases
```

```python
sns.set(rc={'figure.figsize':(15,10)})
sns.barplot(x='state',y='cured',data=top_cured_cases,hue='state')
plt.show()
```
